# Exercício 2
Uma editora estuda a possibilidade de lançar novamente as publicações
Helena, Senhora e A Moreninha. Para isto, efetuou uma pesquisa de mercado e concluiu
que em cada 1000 pessoas consultadas: 600 leram A Moreninha; 400 leram Helena; 300
leram Senhora; 200 leram A Moreninha e Helena; 150 leram A Moreninha e Senhora;
100 leram Senhora e Helena; 20 leram as três obras; Calcule:

*   A) O número de pessoas que leu apenas uma das obras.
*   B) O número de pessoas que não leu nenhuma das três obras.
*   C) O número de pessoas que leu duas ou mais obras.









## Diagrama

```python
U = 1000 # Total de pessoas consultadas
M = 600 # Leram a Moreninha
H = 400 # Leram Helena
S = 300 # Leram Senhora
MH = 200 # Leram Moreninha e Helena
MS = 150 # Leram Moreninha e Senhora
SH = 100 # Leram Senhora e Helena
MHS = 20 #Leram Moreninha, Helena e Senhora

venn = venn3(subsets=(M, H, MH, S, MS, SH, MHS),
             set_labels=('Moreninha','Helena','Senhora'))

venn.get_patch_by_id('100').set_edgecolor('Black')
venn.get_patch_by_id('100').set_lw(2)
venn.get_patch_by_id('010').set_edgecolor('black')
venn.get_patch_by_id('010').set_lw(2)
venn.get_patch_by_id('001').set_edgecolor('black')
venn.get_patch_by_id('001').set_lw(2)
venn.get_patch_by_id('110').set_edgecolor('black')
venn.get_patch_by_id('110').set_lw(2)
venn.get_patch_by_id('101').set_edgecolor('black')
venn.get_patch_by_id('101').set_lw(2)
venn.get_patch_by_id('011').set_edgecolor('black')
venn.get_patch_by_id('011').set_lw(2)
venn.get_patch_by_id('111').set_edgecolor('black')
venn.get_patch_by_id('111').set_lw(2)

plt.title('Pessoas Consultadas')
plt.show()
```

## A) O número de pessoas que leu apenas uma das obras.

```python
print("Moreninha:", (M - MH - MS - MHS))
print("Helena:", (H - MH - SH - MHS))
print("Senhora:", (S - MS - SH - MHS))
```

## B) O número de pessoas que não leu nenhuma das três obras.

```python
print("Total que não leram:", U - (M + (H - MH - SH - MHS) + (S - MS - SH - MHS)))
```

## C) O número de pessoas que leu duas ou mais obras.

```python
print("Total que leram 2 ou mais:", MH + (SH - MHS) + (MS - MHS))
```

#Exercício 4
Uma pesquisa de mercado sobre a preferência de 200 consumidores por três produtos P1, P2 e P3, mostrou que, dos entrevistados, 20 consumiam os três produtos;
30 os produtos P1 e P2; 50 os produtos P2 e P3; 60 os produtos P1 e P3; 120 o produto P1; 75 o produto P2. Se todas as 200 pessoas entrevistadas deram preferência por pelo menos um dos produtos, quantas pessoas consumiram:

*   A) somente o produto P3?
*   B) Pelo menos dois dos produtos, ou seja, (P1 e P2) ou (P1 e P3) ou (P2 e P3) ou (P1 e P2 e P3)?
*   C) Os produtos P1 e P2 e não P3?





## Diagrama

```python
U = 200
P1 = 120 # Consumia o Produto 1
P2 = 75 # Consumia o Produto 2
P3 = 0
P1P2 = 30 # Consumia o Produto 1 e 2
P2P3 = 50 # Consumia o Produto 2 e 3
P1P3 = 60 # Consumia o Produto 1 e 3
P1P2P3 = 20 # Consumia os 3 produtos


venn = venn3(subsets=(P1, P2, P1P2, P3, P1P3, P2P3, P1P2P3),
             set_labels=('Produto 1','Produto 2', 'Produto 3'))

venn.get_label_by_id('001').set_text('?')

plt.title('Total de Consumidores: 200')
plt.show()
```

##A) somente o produto P3?

```python
print('Total:', U - (P1 + (P2 - P1P2 - P1P2P3) ))
```

##B) Pelo menos dois dos produtos, ou seja, (P1 e P2) ou (P1 e P3) ou (P2 e P3) ou (P1 e P2 e P3)?

```python
print('Total:', P1P2 + (P1P3 - P1P2P3) + (P2P3 - P1P2P3))
```

##C) Os produtos P1 e P2 e não P3?

```python
print('Total:', P1P2 - P1P2P3)
```

problemas de som e 3.500 não apresentavam nenhum dos tipos de problema citados. Então o número de aparelhos que apresentavam somente problemas de imagem é:

*   a) 4 000
*   b) 3 700
*   c) 3 500
*   d) 2 800
*   e) 2 500








##Diagrama

```python
U = 10000
imagem = 4000 # Acertaram problema 1
som = 2800 # Acertaram ambos
sem_problema = 3500 # Acertaram problema 2
imagem_e_som = -(U - (imagem + som + sem_problema))


venn = venn2(subsets=(imagem, som, imagem_e_som),
             set_labels=('Problemas de Imagem','Problemas de Som'))

plt.title('Total de Pessoas: 10.000\nSem problemas: 3.500')
plt.show()
```

```python
print('Resposta:', imagem - imagem_e_som)
```
